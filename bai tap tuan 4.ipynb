```python
#W4A8
while True:
    try:
        n = int(input("Nhập vào một số nguyên dương n: "))
        if n > 0:
            break
        else:
            print("Lỗi: Vui lòng nhập số nguyên dương (n > 0).")
    except ValueError:
        print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
so_buoc = 0
so_hien_tai = n 

print(f"\nBắt đầu với n = {so_hien_tai}")
print("")
while True:
    so_tam_kiem_tra = so_hien_tai
    so_dao_nguoc = 0
    while so_tam_kiem_tra > 0:
        chu_so = so_tam_kiem_tra % 10
        so_dao_nguoc = so_dao_nguoc * 10 + chu_so
        so_tam_kiem_tra //= 10
    if so_hien_tai == so_dao_nguoc:
        break 
    so_tiep_theo = so_hien_tai + so_dao_nguoc
    so_buoc += 1
    print(f"Bước {so_buoc}: {so_hien_tai} + {so_dao_nguoc} = {so_tiep_theo}")
    so_hien_tai = so_tiep_theo
print("")
print(f" Kết quả:")
print(f"Số bước cần thiết: {so_buoc}")
print(f"Giá trị Palindrome thu được: {so_hien_tai}")
```

```python
#W4A10
n = int(input())
x_dai_nhat = 1      
L_lon_nhat = 1         
for x in range(1, n + 1):
    do_dai_hien_tai = 1
    hien_tai = x
    while hien_tai > 1:
        if hien_tai % 2 == 0:
            hien_tai //= 2
        else:
            hien_tai = 3 * hien_tai + 1
        do_dai_hien_tai += 1
    if do_dai_hien_tai > L_lon_nhat:
        L_lon_nhat = do_dai_hien_tai
        x_dai_nhat = x
print(f"Số có độ dài chuỗi Collatz lớn nhất (x): {x_dai_nhat}")
print(f"Độ dài chuỗi tương ứng (L): {L_lon_nhat}")
```
