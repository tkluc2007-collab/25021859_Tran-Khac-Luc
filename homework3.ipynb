```python
#Bài 14
print("Nhập các số (dừng khi nhập -1):")
try:
    n = int(input("Nhập số: "))
    if n == -1:
        min_val = None
    else:
        min_val = n
        max_val = n
except ValueError:
    print("Lỗi nhập liệu.")
    min_val = None
while min_val is not None:
    try:
        n = int(input("Nhập số: "))
    except ValueError:
        print("Lỗi nhập liệu. Bỏ qua.")
        continue

    if n == -1:
        break

    if n > max_val:
        max_val = n
    if n < min_val:
        min_val = n
if min_val is not None:
    print(f"Số lớn nhất (Max): {max_val}")
    print(f"Số nhỏ nhất (Min): {min_val}")
else:
    print("Không có số hợp lệ nào được nhập.")
```

```python
#W4A19
def max_fib_under(A):
    f1, f2 = 1, 1
    while f2 <= A:
        f1, f2 = f2, f1 + f2
    return f1

A = int(input("Nhập A: "))
print("Kết quả:", max_fib_under(A))

    
```
