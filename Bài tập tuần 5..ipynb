```python
#W5A1
def Max_of_two(a,b):
    if a > b:
        return a
    else:
        return b
a , b= map(int,input().split())
print(Max_of_two(a,b))
```

```python
#W5A3
def so_nto(n):
    if n <= 1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
n = int(input())
if so_nto(n):
    print("n là số nto")
else:
    print("n không là số nto")
        
        
    
```

```python
#W5A4
def prime_number(n):
    tong_uoc = 0
    if n < 1:
        return False
    for i in range(1,int(n/2)+1):
        if n%i ==0:
            tong_uoc += i
    if tong_uoc == n:
        return True
n = int(input())
if prime_number(n):
    print("n là số hoàn hảo")
else:
    print("n không là số hoàn hảo")
            
```

```python
#W5A5
def tim_so_dau_tien(lst,k):
    for i,v in enumerate(lst):
        if v == k:
            return i
    return -1
lst=map(int,input().split())
k = int(input())
print(tim_so_dau_tien(lst,k))
```

```python
#W5A6
def giai_thua(n):
    if n==0:
        return 0
    if n < 0:
        return -1
    result = 1
    for i in range(1,n+1):
        result *= i
    return result
n = int(input())
print(giai_thua(n))
        
```

```python
#W5A7
def simple_calc(num1, num2, op):
    if op == '+':
        return num1 + num2
    elif op == '-':
        return num1 - num2
    elif op == '*':
        return num1 * num2
    elif op == '/':
        if num2 == 0:
            return "Lỗi: chia cho 0!"
        return num1 / num2
    else:
        return "Toán tử không hợp lệ!"
a = float(input())
b = float(input())
op = input("Nhập toán tử (+, -, *, /): ")
print("Kết quả:", simple_calc(a, b, op))
```

```python
#W5A8
def hamming_distance(x, y):
    return bin(x ^ y).count("1")
x = int(input())
y = int(input())
print(hamming_distance(x, y))

```

```python
#W5A9
def sum_of_digits(n):
    return sum(int(d) for d in str(abs(n)))
n = int(input())
print(sum_of_digits(n))
```

```python
#W5A10
def is_anagram(a, b):
    return sorted(a) == sorted(b)
a = input().strip()
b = input().strip()
print(is_anagram(a, b))
```
